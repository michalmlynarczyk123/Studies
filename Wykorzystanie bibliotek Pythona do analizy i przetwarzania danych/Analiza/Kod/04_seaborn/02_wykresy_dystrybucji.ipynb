```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips, x='tip')
```

```python
_ = sns.displot(tips, x='tip', bins=30)
```

### Joinplot

```python
_ = sns.jointplot(tips, x='total_bill', y='tip')
```

```python
_ = sns.jointplot(tips, x='total_bill', y='tip', hue='smoker')
```

```python
_ = sns.pairplot(tips)
```

```python
_ = sns.pairplot(tips, hue='time')
```
