```python
import pandas as pd
import seaborn as sns
```

```python
df_houses_raw = pd.read_csv('../../Pliki/04_seaborn/house_price_filter.csv')
df_houses_raw.head()
```

```python
df_houses = df_houses_raw.copy()
```

```python
df_houses.info()
```

### Analiza jednoczynnikowa

```python
_ = sns.catplot(df_houses, x='year', kind='count')
```

```python
_ = sns.catplot(df_houses, x='month', kind='count')
```

Histogram ceny

```python
_ = sns.histplot(df_houses, x='price', bins=70)
```

Histogram ceny za m2

```python
_ = sns.histplot(df_houses, x='price_per_meter', bins=70)
```

```python
_ = sns.catplot(df_houses, x='floor', kind='count')
```

Histogram powierzchni

```python
_ = sns.histplot(df_houses, x='area', bins=50)
```

Ilość pokoi w ogłoszeniach

```python
_ = sns.catplot(df_houses, x='rooms', kind='count')
```

Rodzaje ofert

```python
_ = sns.catplot(df_houses, x='offer_type', kind='count')
```

Rodzaj budynku

```python
_ = sns.catplot(df_houses, y='offer_type_of_building', kind='count')
```

Rynek

```python
_ = sns.catplot(df_houses, x='market', kind='count')
```

Województwo

```python
_ = sns.catplot(df_houses, y='voivodeship', kind='count')
```

Miasto

```python
_ = sns.catplot(df_houses, y='city_name', kind='count', order=df_houses.value_counts('city_name').iloc[:20].index)
```
