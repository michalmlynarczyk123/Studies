```python
import numpy as np
import matplotlib.pyplot as plt
```

### Wykres liniowy

```python
_ = plt.plot([0, 1], [0, 1])
```

```python
x = np.arange(-1, 1, 0.1)
y = x ** 2
```

```python
_ = plt.plot(x, y)
```

```python
plt.plot(x, y)
_ = plt.title('Wykres x * y')
```

```python
plt.plot(x, y)
plt.xlabel('Oś x')
_ = plt.ylabel('Oś y')
```

```python
x1 = np.arange(-1, 1, 0.1)
y1 = x1 ** 3
```

```python
plt.plot(x, y)
_ = plt.plot(x1, y1)
```

```python
plt.plot(x, y, label='x ^ 2')
plt.plot(x1, y1, label='x ^ 3')
_ = plt.legend()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.subplot(1, 2, 2)
_ = plt.plot(x1, y1)
```
